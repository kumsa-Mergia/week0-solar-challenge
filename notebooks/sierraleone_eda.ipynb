```python
class SolarDataAnalyzer:
    def __init__(self, input_path, output_path):
        from src.utils import (
            load_data,
            Summary_Statistics,
            missing_report,
            null_value,
            outlier_detection,
            export_cleaned_data,
            time_series_analysis,
            summary_statistics_missing_value,
            correlation_analysis,
            wind_distribution_analysis,
            temperature_analysis,
            bubble_chart
        )
        self.input_path = input_path
        self.output_path = output_path
        self.df = load_data(input_path)
        # Store function references
        self.Summary_Statistics = Summary_Statistics
        self.missing_report = missing_report
        self.null_value = null_value
        self.outlier_detection = outlier_detection
        self.export_cleaned_data = export_cleaned_data
        self.time_series_analysis = time_series_analysis
        self.summary_statistics_missing_value = summary_statistics_missing_value
        self.correlation_analysis = correlation_analysis
        self.wind_distribution_analysis = wind_distribution_analysis
        self.temperature_analysis = temperature_analysis
        self.bubble_chart = bubble_chart

    def clean_data(self):
        self.outlier_detection(self.df)
        self.export_cleaned_data(self.df, self.output_path)

    def analyze(self):
        self.Summary_Statistics(self.df)
        self.missing_report(self.df)
        self.null_value(self.df)
        self.time_series_analysis(self.df)
        self.summary_statistics_missing_value(self.df)
        self.correlation_analysis(self.df)
        self.wind_distribution_analysis(self.df)
        self.temperature_analysis(self.df)
        self.bubble_chart(self.df)

    def run_all(self):
        self.clean_dat
```
